```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# step 1: Load the excwl file into a dataframe
df = pd.read_csv('Car_sales.csv')
df.head()
```

```python
# step 2: Extract the necessary data for the bar chart
manufacturer =df['Manufacturer']
sales =df['Sales_in_thousands']
```

```python
# step 3:Create and Customize the bar chart
plt.figure(figsize=(15,10))
plt.barh(manufacturer, sales)
plt.xlabel('car Manufacturers')
plt.ylabel('Sales in 1000s')
plt.title('Car sales')
plt.show()
```
